```python
# I used folders in order: 200_1, 200_2, 100, 200_3, 200_4

import import_ipynb
import light_config
import os
from light_config import *


baseDir = '/Users/denis/Downloads/lfw-deepfunneled200_4' 
targetFile = ''
sep = '/'
outputDir = '/Users/denis/Desktop'


# -------------Step 1, iterate through all folders and all pictures in side folders
index = 0
results = 'Person 1, Person 2, Expected, Result, Percent identity \n'

for folder in os.listdir(baseDir):
    index = index + 1
    if len(os.listdir(baseDir + sep + folder)) > 1:
        count = 1
        for file in os.listdir(baseDir + sep + folder):

#--------------Step 2, select target image, generate vector for it
            if count == 1:  
                try:
                    count = count + 1
                    targetFile = file
                    targetVector = get_embeddings_l_vgg16([baseDir + '/' + folder + '/' + file])
                except Exception as err:
                    print("---------Error-----------")
                    
#--------------Step 3, select next image, generate vector for it
            else:
                count = count + 1
                path = baseDir + '/' + folder + '/' + file
                try:
                    testVector = get_embeddings_l_vgg16([path])
                    result = is_match_mass(targetVector, testVector)
                    textResult = 'Match'
                    print("\n-----------result---------->", targetFile, "/", file, "===>", result, index)
                    if result <= 50 :
                        textResult = 'Not Match'
                    results = results + targetFile + ',' + file + ',' + 'Match' +  ',' + textResult + ',' + str(result) + '\n'
                except Exception as err:
                    print("---------Error-----------")
    
    if index >= 200:
        break
        
print('\n\n Writing output file...')
output = open(outputDir + '/results_match_vgg16_200_4.csv', 'w')
output.write(results)
output.close()
```
