```python
# I used folders in order: 200_1, 200_2, 100, 200_3, 200_4

import import_ipynb
import light_config
import os
from light_config import *


baseDir = '/Users/denis/Downloads/lfw-deepfunneled200_1' 
targetFile = 'Aaron_Eckhart_0001.jpg'
target = baseDir + '/Aaron_Eckhart/' + targetFile
sep = '/'
outputDir = '/Users/denis/Desktop'

# Step 1, select target image, generate vector for it
targetVector = get_embeddings([target])

# Step 2, iterate through all folders and all pictures in side folders

results = 'Person 1, Person 2, Expected, Result, Percent identity \n'

index = 0
for folder in os.listdir(baseDir):
    for file in os.listdir(baseDir + sep + folder):
        path = baseDir + '/' + folder + '/' + file
        try:
            testVector = get_embeddings([path])
            result = is_match_mass(targetVector, testVector)
            textResult = 'Match'
            print("\n--------------------->", result, index)
            if result <= 50 :
                textResult = 'Not Match'
            results = results + targetFile + ',' + file + ',' + 'Not match' +  ',' + textResult + ',' + str(result) + '\n'
        except Exception as err:
            print("---------multiple faces detected-----------")

    index = index + 1
    if index > 200:
        break
        
print('\n\n Writing output file...')
output = open(outputDir + '/results200_1.csv', 'w')
output.write(results)
output.close()
```
